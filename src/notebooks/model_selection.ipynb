```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("../../data/features/features_AAPL.csv")
df.head()
```

```python
if "Date" in df.columns:
    df = df.drop(columns=["Date"])

X = df.drop(columns=["target"])
corr_matrix = X.corr()
print(corr_matrix)

plt.figure(figsize=(14, 12))
sns.heatmap(
    corr_matrix,
    cmap="coolwarm",
    center=0,
    square=True,
    cbar_kws={"shrink": 0.8}
)
plt.title("Heatmapa korelacji cech")
plt.tight_layout()
plt.show()
```
